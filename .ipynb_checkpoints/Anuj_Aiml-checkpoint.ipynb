```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')
le = LabelEncoder()
anuj_df['Country'] = le.fit_transform(anuj_df['Country'])
anuj_df['City'] = le.fit_transform(anuj_df['City'])
anuj_df['Region'] = le.fit_transform(anuj_df['Region'])
```

```python
features = anuj_df.drop(columns=['AirQuality'])
labels = anuj_df['AirQuality']
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_features, labels, test_size=0.2, random_state=42)
```

```python
rf = RandomForestRegressor(random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
```

```python
print('MAE:', mean_absolute_error(y_test, y_pred))
print('MSE:', mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(mean_squared_error(y_test, y_pred)))
print('R^2 Score:', r2_score(y_test, y_pred))
```

```python
pollution_anuj_df = anuj_df[['AirQuality', 'WaterPollution']]
kmeans = KMeans(n_clusters=3, random_state=42)
anuj_df['Cluster'] = kmeans.fit_predict(pollution_anuj_df)

# Correct argument name here:
sns.scatterplot(x='AirQuality', y='WaterPollution', hue='Cluster', data=anuj_df)
plt.title('City Pollution Clusters')
plt.show()
```
