```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

anuj_df = anuj_df.drop_duplicates()

# Fill missing values in columns, avoiding chained assignments
for col in anuj_df.columns:
    if col != 'Hookload':  # Exclude the specified column
        if anuj_df[col].isnull().any():
            mean_val = anuj_df[col].mean()
            anuj_df[col] = anuj_df[col].fillna(mean_val)  # Use assignment instead of inplace

# Create a list of columns to scale
cols_to_scale = [col for col in anuj_df.columns if col != 'Hookload']

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Filter out columns that have missing values or insufficient unique values
valid_cols_to_scale = [
    col for col in cols_to_scale if anuj_df[col].notna().all() and anuj_df[col].nunique() > 1
]

# Print warnings for excluded columns
excluded_cols = set(cols_to_scale) - set(valid_cols_to_scale)
if excluded_cols:
    print(f"Warning: These columns were skipped due to NaNs or insufficient unique values: {excluded_cols}")

# Proceed to scale only the valid columns
anuj_df[valid_cols_to_scale] = scaler.fit_transform(anuj_df[valid_cols_to_scale])

# Verify the transformation
print(anuj_df.head())
```

# SECTION 4: (Perform outlier removal, apply Principal Component Analysis (PCA)), LDA, t-SNE
 ***Outlier removal***
 * Z-Score Method (Standardization) Roll No: EVEN
 * Interquartile Range (IQR) Method Roll No: ODD
 
 ***Feature Extraction***
* Perform PCA, LDA, TSNE with 3 components.
* Print all necessary for PCA results.
* Plot new PC1 to PC3
* Choose **Hookload** for RO_data, **Hookload** for Drilling_data and **T_o** for Flocculant_data as your target variable
* Generate a Scree Plot for PCA, LDA.

```python
print(anuj_df.shape)

def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_no_outliers = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return df_no_outliers

for col in anuj_df.columns:
    anuj_df = remove_outliers_iqr(anuj_df, col)

print(anuj_df.shape)
```

```python
X = anuj_df.drop('Hookload', axis=1)
y = anuj_df['Hookload']


pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)

print("PCA Explained Variance Ratio:", pca.explained_variance_ratio_)
print("PCA Singular Values:", pca.singular_values_)

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA - PC1 vs PC2')
plt.colorbar(label='Hookload')
plt.show()


explained_variance = pca.explained_variance_ratio_
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, marker='o')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot for PCA')
plt.show()
```

```python



lda = LDA(n_components=3)
num_bins = 8  # Change this based on the number of classes you need
y_binned = pd.cut(y, bins=num_bins, labels=False)  # Discretize continuous values
X_lda = lda.fit_transform(X, y_binned)

print("LDA Explained Variance Ratio:", lda.explained_variance_ratio_)

plt.figure(figsize=(8, 6))
plt.scatter(X_lda[:, 0], X_lda[:, 1], c=y, cmap='viridis')
plt.xlabel('LD1')
plt.ylabel('LD2')
plt.title('LDA - LD1 vs LD2')
plt.colorbar(label='Hookload')
plt.show()

explained_variance_lda = lda.explained_variance_ratio_
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(explained_variance_lda) + 1), explained_variance_lda, marker='o')
plt.xlabel('Linear Discriminant')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot for LDA')
plt.show()
```

```python
import matplotlib.pyplot as plt
tsne = TSNE(n_components=3, random_state=42)
X_tsne = tsne.fit_transform(X)


plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.title('t-SNE Visualization')
plt.colorbar(label='Hookload')
plt.show()
```

# SECTION 6: Perform Hyperparameter tuning using GridSearchCV

* Choose your own parameters to tune **(Don't take too many variations as time increases drastically)**
* Perform Hyperparameter tuning using GridSearchCV
* Retrain the classifiers using optimal hyperparameters
* Calculate the new metrics and plot the decision boundary again

```python
# Define the parameter grid for KNN
param_grid = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

# Perform GridSearchCV for KNN on the original data
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.25, random_state=42
)
knn = KNeighborsClassifier()
grid_search = GridSearchCV(estimator=knn, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)

# Get the best parameters and retrain the classifier
best_params = grid_search.best_params_
print("Best Parameters for KNN (Original Data):", best_params)
knn_best = KNeighborsClassifier(**best_params)
knn_best.fit(X_train, y_train)
y_pred = knn_best.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy for KNN (Original Data - Tuned): {accuracy:.4f}")
print("\nClassification Report for KNN (Original Data - Tuned):")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix for KNN (Original Data - Tuned):")
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix for KNN (Original Data - Tuned)")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# Perform GridSearchCV and retrain for PCA, LDA, and t-SNE data
for name, X_transformed in zip(["PCA", "LDA", "t-SNE"], [X_pca[:, :2], X_lda[:, :2], X_tsne]):
    X_train_t, X_test_t, y_train_t, y_test_t = train_test_split(
        X_transformed, y_encoded, test_size=0.25, random_state=42
    )

    knn_t = KNeighborsClassifier()
    grid_search_t = GridSearchCV(estimator=knn_t, param_grid=param_grid, cv=5)
    grid_search_t.fit(X_train_t, y_train_t)

    best_params_t = grid_search_t.best_params_
    print(f"\nBest Parameters for KNN ({name} Data):", best_params_t)

    knn_best_t = KNeighborsClassifier(**best_params_t)
    knn_best_t.fit(X_train_t, y_train_t)
    y_pred_t = knn_best_t.predict(X_test_t)

    accuracy_t = accuracy_score(y_test_t, y_pred_t)
    print(f"\nAccuracy for KNN ({name} Data - Tuned): {accuracy_t:.4f}")
    print(f"\nClassification Report for KNN ({name} Data - Tuned):")
    print(classification_report(y_test_t, y_pred_t))

    print(f"\nConfusion Matrix for KNN ({name} Data - Tuned):")
    cm_t = confusion_matrix(y_test_t, y_pred_t)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm_t, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Confusion Matrix for KNN ({name} Data - Tuned)")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
```
