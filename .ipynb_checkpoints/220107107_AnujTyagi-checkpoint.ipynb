# SECTION 3
 Preprocessing
* Remove duplicates if they are present
* Fill null values with the mean(Roll no: Even)/median (Roll no: ODD) if any are found
  
  ***Feature scaling:***
* Min-Max Scaling (Normalization) Roll No: Even
* Standardization (Z-Score Normalization) Roll No: ODD

  ***Correlation Heatmap***
 * Drop the columns based on the heat map if required

```python
anuj_df = anuj_df.drop_duplicates()

for col in anuj_df.columns:
    if anuj_df[col].isnull().any():
        median_val = anuj_df[col].median()
        anuj_df[col].fillna(median_val, inplace=True)

for col in anuj_df.select_dtypes(include=np.number).columns:
    anuj_df[col] = (anuj_df[col] - anuj_df[col].mean()) / anuj_df[col].std()
```

```python

plt.figure(figsize=(12, 10))
sns.heatmap(anuj_df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

# SECTION 4: (apply Principal Component Analysis (PCA), LDA, t-SNE without removing outliers)
 ***Feature Extraction***

```python


from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt


X = anuj_df.drop('ROP', axis=1)
y = anuj_df['ROP']


pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)

print("Explained Variance Ratio:", pca.explained_variance_ratio_)
print("Singular Values:", pca.singular_values_)
print("Components:", pca.components_)


plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA: PC1 vs PC2')
plt.colorbar(label='ROP')
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 2], c=y, cmap='viridis')
plt.xlabel('PC1')
plt.ylabel('PC3')
plt.title('PCA: PC1 vs PC3')
plt.colorbar(label='ROP')
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 1], X_pca[:, 2], c=y, cmap='viridis')
plt.xlabel('PC2')
plt.ylabel('PC3')
plt.title('PCA: PC2 vs PC3')
plt.colorbar(label='ROP')
plt.show()


plt.figure(figsize=(8, 6))
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1), pca.explained_variance_ratio_, marker='o', linestyle='--')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot for PCA')
plt.show()


y = pd.cut(anuj_df['ROP'], bins=3, labels=False)


lda = LDA(n_components=2)
X_lda = lda.fit_transform(X, y)



plt.figure(figsize=(8, 6))
plt.scatter(X_lda[:, 0], X_lda[:, 1], c=y, cmap='viridis')
plt.xlabel('LD1')
plt.ylabel('LD2')
plt.title('LDA')
plt.colorbar(label='ROP')
plt.show()

plt.figure(figsize=(8, 6))
plt.plot(range(1, lda.explained_variance_ratio_.shape[0] + 1), lda.explained_variance_ratio_, marker='o', linestyle='--')
plt.xlabel('Number of Linear Discriminants')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot for LDA')
plt.show()



tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.title('t-SNE')
plt.colorbar(label='ROP')
plt.show()

```

# SECTION 5: (Perform outlier removal, and apply Principal Component Analysis (PCA)), LDA, t-SNE
 ***Outlier removal***
 * Z-Score Method (Standardization) Roll No: Even
 * Interquartile Range (IQR) Method Roll No: ODD

 ***Feature Extraction***
* Perform PCA.
* Print all necessary for PCA results.
* Plot new PC1,PC2,PC3
* Generate a Scree Plot for PCA, LDA.
* Scatter plot for t-SNE

```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn.manifold import TSNE
from scipy import stats

z = np.abs(stats.zscore(anuj_df.select_dtypes(include=np.number)))
anuj_df_o = anuj_df[(z < 3).all(axis=1)]

X_o = anuj_df_o.drop('ROP', axis=1)
y_o = anuj_df_o['ROP']

pca_o = PCA(n_components=3)
X_pca_o = pca_o.fit_transform(X_o)

print("\nExplained Variance Ratio (after outlier removal):", pca_o.explained_variance_ratio_)
print("Singular Values (after outlier removal):", pca_o.singular_values_)
print("Components (after outlier removal):", pca_o.components_)

plt.figure(figsize=(8, 6))
plt.scatter(X_pca_o[:, 0], X_pca_o[:, 1], c=y_o, cmap='viridis')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA (after outlier removal): PC1 vs PC2')
plt.colorbar(label='ROP')
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(X_pca_o[:, 0], X_pca_o[:, 2], c=y_o, cmap='viridis')
plt.xlabel('PC1')
plt.ylabel('PC3')
plt.title('PCA (after outlier removal): PC1 vs PC3')
plt.colorbar(label='ROP')
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(X_pca_o[:, 1], X_pca_o[:, 2], c=y_o, cmap='viridis')
plt.xlabel('PC2')
plt.ylabel('PC3')
plt.title('PCA (after outlier removal): PC2 vs PC3')
plt.colorbar(label='ROP')
plt.show()

plt.figure(figsize=(8, 6))
plt.plot(range(1, len(pca_o.explained_variance_ratio_) + 1), pca_o.explained_variance_ratio_, marker='o', linestyle='--')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot for PCA (after outlier removal)')
plt.show()


y_o = pd.cut(anuj_df_o['ROP'], bins=3, labels=False)
lda_o = LDA(n_components=2)
X_lda_o = lda_o.fit_transform(X_o, y_o)

plt.figure(figsize=(8, 6))
plt.scatter(X_lda_o[:, 0], X_lda_o[:, 1], c=y_o, cmap='viridis')
plt.xlabel('LD1')
plt.ylabel('LD2')
plt.title('LDA (after outlier removal)')
plt.colorbar(label='ROP')
plt.show()


plt.figure(figsize=(8, 6))
plt.plot(range(1, lda_o.explained_variance_ratio_.shape[0] + 1), lda_o.explained_variance_ratio_, marker='o', linestyle='--')
plt.xlabel('Number of Linear Discriminants')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot for LDA (after outlier removal)')
plt.show()


tsne_o = TSNE(n_components=2, random_state=42)
X_tsne_o = tsne_o.fit_transform(X_o)

plt.figure(figsize=(8, 6))
plt.scatter(X_tsne_o[:, 0], X_tsne_o[:, 1], c=y_o, cmap='viridis')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.title('t-SNE (after outlier removal)')
plt.colorbar(label='ROP')
plt.show()
```

# SECTION 6: Perform outlier removal, feature selection, and apply Principal Component Analysis (PCA), LDA, t-SNE
***Outlier removal***
 * Z-Score Method (Standardization) Roll No: Even
 * Interquartile Range (IQR) Method Roll No: ODD

 ***Feature selection***
* Perform Feature Selection using Lasso Regression(L1 approach) Roll No: Even
* Perform Feature Selection using Ridge Regression(L2 approach) Roll No: ODD

  ***Feature Extraction***
* Perform PCA.
* Print all necessary for PCA results.
* Plot new PC1,PC2,PC3
* Generate a Scree Plot for PCA, LDA.
* Scatter plot for t-SNE

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import Ridge


ridge = Ridge(alpha=1.0)
ridge.fit(X_o, y_o)

feature_importances = abs(ridge.coef_)


sorted_idx = np.argsort(feature_importances)[::-1]

top_n_features = 10
selected_features = X_o.columns[sorted_idx[:top_n_features]]
X_fs = X_o[selected_features]


pca_fs = PCA(n_components=3)
X_pca_fs = pca_fs.fit_transform(X_fs)

print("\nExplained Variance Ratio (after feature selection):", pca_fs.explained_variance_ratio_)
print("Singular Values (after feature selection):", pca_fs.singular_values_)
print("Components (after feature selection):", pca_fs.components_)


plt.figure(figsize=(8, 6))
plt.scatter(X_pca_fs[:, 0], X_pca_fs[:, 1], c=y_o, cmap='viridis')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA (after feature selection): PC1 vs PC2')
plt.colorbar(label='ROP')
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(X_pca_fs[:, 0], X_pca_fs[:, 2], c=y_o, cmap='viridis')
plt.xlabel('PC1')
plt.ylabel('PC3')
plt.title('PCA (after feature selection): PC1 vs PC3')
plt.colorbar(label='ROP')
plt.show()

plt.figure(figsize=(8, 6))
plt.scatter(X_pca_fs[:, 1], X_pca_fs[:, 2], c=y_o, cmap='viridis')
plt.xlabel('PC2')
plt.ylabel('PC3')
plt.title('PCA (after feature selection): PC2 vs PC3')
plt.colorbar(label='ROP')
plt.show()

plt.figure(figsize=(8, 6))
plt.plot(range(1, len(pca_fs.explained_variance_ratio_) + 1), pca_fs.explained_variance_ratio_, marker='o', linestyle='--')
plt.xlabel('Number of Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot for PCA (after feature selection)')
plt.show()

lda_fs = LDA(n_components=2)
X_lda_fs = lda_fs.fit_transform(X_fs, y_o)

plt.figure(figsize=(8, 6))
plt.scatter(X_lda_fs[:, 0], X_lda_fs[:, 1], c=y_o, cmap='viridis')
plt.xlabel('LD1')
plt.ylabel('LD2')
plt.title('LDA (after feature selection)')
plt.colorbar(label='ROP')
plt.show()

plt.figure(figsize=(8, 6))
plt.plot(range(1, lda_fs.explained_variance_ratio_.shape[0] + 1), lda_fs.explained_variance_ratio_, marker='o', linestyle='--')
plt.xlabel('Number of Linear Discriminants')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot for LDA (after feature selection)')
plt.show()

tsne_fs = TSNE(n_components=2, random_state=42)
X_tsne_fs = tsne_fs.fit_transform(X_fs)

plt.figure(figsize=(8, 6))
plt.scatter(X_tsne_fs[:, 0], X_tsne_fs[:, 1], c=y_o, cmap='viridis')
plt.xlabel('t-SNE Dimension 1')
plt.ylabel('t-SNE Dimension 2')
plt.title('t-SNE (after feature selection)')
plt.colorbar(label='ROP')
plt.show()
```

Observations on Sections 4, 5, and 6

Section 4 (PCA, LDA, t-SNE without outlier removal):
PCA successfully reduced dimensionality while maintaining variance, although the clustering was not clearly defined. LDA enhanced class separation based on the Rate of Prediction (ROP), but its effectiveness was moderate. t-SNE highlighted data clusters, yet was computationally intensive.

Section 5 (Outlier removal + PCA, LDA, t-SNE):

Although this increased computational time, the enhanced separation indicated that outliers had previously obscured the genuine structures of the data.

Section 6 (Outlier removal, feature selection + PCA, LDA, t-SNE):
Incorporating feature selection through Ridge regression not only reduced dimensionality but also improved computational efficiency, likely enhancing model performance. The visualizations from PCA, LDA, and t-SNE presented clearer patterns due to the selection of essential features, although the trade-offs in accuracy depended on which features were excluded.

Summary:
Overall, outlier removal consistently enhanced the clarity of visualizations and had the potential to improve model performance. Feature selection via regularization increased efficiency while retaining crucial information. PCA is most effective for preserving variance, LDA excels in class separability, and t-SNE is valuable for visualizing high-dimensional data.
