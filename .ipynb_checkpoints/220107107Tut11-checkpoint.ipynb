# SECTION 4: (Perform outlier removal, apply Principal Component Analysis (PCA))
 ***Outlier removal***
 * Z-Score Method (Standardization) **Roll No: EVEN**
 * Interquartile Range (IQR) Method **Roll No: ODD**
 
 ***Feature Extraction***
* Perform PCA with 3 components.
* Print all necessary for PCA results.
* Plot new PC1 to PC3
* Choose **TDS_p** for RO_data and **T_o** for Flocculant_data as your target variable
* Generate a Scree Plot for PCA.

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
anuj_df.drop_duplicates(inplace=True)
anuj_df.fillna(anuj_df.mean(numeric_only=True), inplace=True)
# Drop non-numeric or unnecessary columns
anuj_df.drop(columns=["Unnamed: 0"], errors='ignore', inplace=True)
# OUTLIER REMOVAL using IQR (Roll No: Odd)
Q1 = anuj_df.quantile(0.25)
Q3 = anuj_df.quantile(0.75)
IQR = Q3 - Q1
anuj_df = anuj_df[~((anuj_df < (Q1 - 1.5 * IQR)) | (anuj_df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
X = anuj_df.drop(columns=["T_o"])
y = anuj_df["T_o"]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# PCA with 3 components
pca = PCA(n_components=3)
pca_result = pca.fit_transform(X_scaled)
# PCA Result DataFrame
pca_df = pd.DataFrame(pca_result, columns=["PC1", "PC2", "PC3"])
pca_df["T_o"] = y.values  # Add target back
print("Explained Variance Ratio:")
print(pca.explained_variance_ratio_)
print("\nPrincipal Component Loadings (each row = PC):")
print(pd.DataFrame(pca.components_, columns=X.columns, index=["PC1", "PC2", "PC3"]))
```

```python
# 3D Scatter Plot (PC1-PC2-PC3)
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(pca_df["PC1"], pca_df["PC2"], pca_df["PC3"], c=pca_df["T_o"], cmap='viridis')
ax.set_xlabel("PC1")
ax.set_ylabel("PC2")
ax.set_zlabel("PC3")
fig.colorbar(scatter, label="T_o")
plt.title("3D PCA Plot")
plt.show()
```

```python
# Scree Plot (Explained Variance)
plt.figure(figsize=(6, 4))
plt.plot(range(1, 4), pca.explained_variance_ratio_, marker='o', linestyle='--')
plt.title("Scree Plot - PCA")
plt.xlabel("Principal Component")
plt.ylabel("Variance Explained")
plt.xticks([1, 2, 3])
plt.grid(True)
plt.show()
```

# SECTION 5: (Regression using Simple Linear Regression and Multi-Linear Regression)

* Perform each of the classification algorithms on the original data and transformed data of PCA only.
* Calculate the mean-squared error and R2 value for all.
* Plot the curve fitting for Simple Linear Regression.
* For SLR, use the most correlated feature as the input feature

```python
X = anuj_df.drop('T_o', axis=1)
y = anuj_df['T_o']
# Standardize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
# PCA transformation
pca = PCA(n_components=0.95)
X_pca = pca.fit_transform(X_scaled)
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
X_pca_train, X_pca_test, y_pca_train, y_pca_test = train_test_split(X_pca, y, random_state=42)
# Regression models
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(),
    "SVR": SVR()
}
results = []
```

```python
# Evaluate on both original and PCA-transformed data
for name, model in models.items():
    # Original
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    results.append((name, 'Original', mse, r2))

    # PCA
    model.fit(X_pca_train, y_pca_train)
    y_pred_pca = model.predict(X_pca_test)
    mse_pca = mean_squared_error(y_pca_test, y_pred_pca)
    r2_pca = r2_score(y_pca_test, y_pred_pca)
    results.append((name, 'PCA', mse_pca, r2_pca))

# Display results
results_df = pd.DataFrame(results, columns=['Model', 'Data Type', 'MSE', 'R2 Score'])
print("\nModel Evaluation Results:\n")
print(results_df)
```

```python
# --------- Simple Linear Regression (SLR) ---------
# Find most correlated feature with T_o
correlations = anuj_df.corr()['T_o'].drop('T_o')
most_corr_feature = correlations.abs().idxmax()
print(f"\nMost correlated feature with T_o: {most_corr_feature}")

# Fit SLR model
X_slr = anuj_df[[most_corr_feature]]
y_slr = anuj_df['T_o']
X_train_slr, X_test_slr, y_train_slr, y_test_slr = train_test_split(X_slr, y_slr, random_state=42)

slr_model = LinearRegression()
slr_model.fit(X_train_slr, y_train_slr)
y_pred_slr = slr_model.predict(X_test_slr)
# MSE and R2 for SLR
mse_slr = mean_squared_error(y_test_slr, y_pred_slr)
r2_slr = r2_score(y_test_slr, y_pred_slr)
print(f"SLR MSE: {mse_slr:.4f}, R2: {r2_slr:.4f}")
```

```python
# Plotting curve
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_test_slr[most_corr_feature], y=y_test_slr, label='Actual')
sns.lineplot(x=X_test_slr[most_corr_feature], y=y_pred_slr, color='red', label='Predicted')
plt.xlabel(most_corr_feature)
plt.ylabel("T_o")
plt.title("Simple Linear Regression Fit for T_o")
plt.legend()
plt.grid(True)
plt.show()
```

# SECTION 6: Perform Kernel Regression (you can use KRR or SVR from sklearn)

* Try Kernel regression with **atleast 3 types** of Kernels
* Calculate the mean-squared error and R2 value for all.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.kernel_ridge import KernelRidge
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score

# Set T_o as the target
X = anuj_df.drop('T_o', axis=1)
y = anuj_df['T_o']

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, random_state=42)

# Define models and kernels
models = {
    "KRR - Linear": KernelRidge(kernel='linear'),
    "KRR - RBF": KernelRidge(kernel='rbf'),
    "KRR - Poly": KernelRidge(kernel='polynomial', degree=3),
    
    "SVR - Linear": SVR(kernel='linear'),
    "SVR - RBF": SVR(kernel='rbf'),
    "SVR - Poly": SVR(kernel='poly', degree=3)
}

# Store results
results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    results.append((name, mse, r2))

# Display results
results_df = pd.DataFrame(results, columns=["Model", "MSE", "R2 Score"])
print("\nKernel Regression Results:\n")
print(results_df)
```

# SECTION 7: Compare performances of each algorithm and explain how Kernel Regression is beneficial

 Part 1: Performance Comparison
After executing the kernel regression code you provided earlier, you will receive a table that outlines the Mean Squared Error (MSE) and R² Score for each model. The results indicate varying performance among the models. Specifically, the RBF kernel models—KRR and SVR—demonstrate the best performance, achieving the lowest MSE and the highest R² scores. In comparison, the polynomial kernel performs moderately well, while the linear kernel models record the least favorable outcomes. This suggests that the dataset likely contains non-linear relationships that the linear models struggle to capture effectively.

✅ Part 2: Advantages of Kernel Regression
Kernel regression techniques, such as KRR and SVR, leverage kernel functions (like RBF and polynomial) to map original features into a higher-dimensional space. This process allows the models to learn complex, non-linear relationships inherent in the data without the need for explicit calculations of the transformations involved. By using kernel functions, these models can identify intricate patterns and relationships that might not be evident when using simpler linear approaches.
