# SECTION 4: (Perform outlier removal, apply Principal Component Analysis (PCA)) and LDA

 ***Feature Extraction***
* Perform PCA and LDA with 3 components.
* Print all necessary for PCA results.
* Plot new PC1 to PC3
* Choose **TDS_p** for RO_data, **HookLoad** for Drilling_data and **T_o** for Flocculant_data as your target variable

```python
from scipy.stats import zscore
z_scores = np.abs(zscore(anuj_df_scaled.select_dtypes(include=[np.number])))
threshold = 3  # Standard threshold for Z-score
anuj_df_scaled = anuj_df_scaled[(z_scores < threshold).all(axis=1)]
print("Shape after outlier removal:", anuj_df_scaled.shape)
```

```python
z_scores = np.abs(zscore(anuj_df_scaled.select_dtypes(include=[np.number])))
threshold = 3  # Standard threshold for Z-score
anuj_df= anuj_df_scaled[(z_scores < threshold).all(axis=1)]

df=anuj_df.copy()
x = df.drop(columns = ['HOOKLOAD_AVG tonfUS'])
y = df['HOOKLOAD_AVG tonfUS']

pca = PCA(n_components=3)
df_pca = pca.fit_transform(anuj_df)

explained_variance = pca.explained_variance_ratio_
print("Explained Variance Ratio:", explained_variance)

df_pca = pd.DataFrame(df_pca, columns=['PC1', 'PC2', 'PC3'])

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df_pca['PC1'], df_pca['PC2'], df_pca['PC3'], c='blue', edgecolors='k', alpha=0.7)

ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('PCA Result (3 Components)')
plt.show()

plt.figure(figsize=(8, 5))
plt.bar(range(1, len(explained_variance) + 1), explained_variance, color='skyblue', edgecolor='black')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot (Bar Chart)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


y_categorical = pd.qcut(y, q=4, labels=False)

lda = LDA(n_components=3)
X_lda = lda.fit_transform(x, y_categorical)

explained_variance_ratio_lda = lda.explained_variance_ratio_
components_lda = np.arange(1, len(explained_variance_ratio_lda) + 1)

plt.figure(figsize=(8, 6))
plt.bar(components_lda, explained_variance_ratio_lda, alpha=0.7, label='Individual explained variance')
plt.plot(components_lda, np.cumsum(explained_variance_ratio_lda), marker='o', color='red', label='Cumulative explained variance')

plt.xlabel('LDA Components')
plt.ylabel('Explained Variance Ratio')
plt.title('LDA Scree Plot')
plt.xticks(components_lda)
plt.legend()
plt.show()
```

# SECTION 5: (Classification using NB, SVM, KNN, LogReg, Decision Tree, and Random Forest)

* Perform each of the classification algorithms on the original data and transformed data of PCA and LDA only
* Print classification report
* Print confusion matrix(Please make the confusion matrix legible.)
* Plot the Decision Boundary Visualization for the KNN and the tree for Decision Tree.

 ***Ensemble Implementation***
* Stacking Classifier Method **Roll No: ODD**
* Voing Classifier Method **Roll No: Even**
* *Optional*: You can try other ensemble algorithms like XGBoost, CatBoost, and AdaBoost

```python
df = anuj_df.copy()

target = 'HOOKLOAD_AVG tonfUS' 
X = df.drop(columns=[target]).values
y = df[target].values

y_binned = pd.qcut(y, q=3, labels=["Low", "Medium", "High"])  # Create 3 categories
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y_binned)

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)


pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)
X_pca_train, X_pca_test = train_test_split(X_pca, test_size=0.2, random_state=42)


lda = LDA(n_components=2)
  # Encode categorical target
X_lda = lda.fit_transform(X, y_encoded)
X_lda_train, X_lda_test, y_lda_train, y_lda_test = train_test_split(X_lda, y_encoded, test_size=0.2, random_state=42)
```

```python
# Define classifiers
classifiers = {
    "Naïve Bayes": GaussianNB(),
    "SVM": SVC(),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "Logistic Regression": LogisticRegression(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier()
}

# Function to Train and Evaluate Model
def evaluate_model(model, X_train, X_test, y_train, y_test, name):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    # Print Classification Report
    print(f"\n🔹 Classification Report for {name}:\n")
    print(classification_report(y_test, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot(cmap='Blues', values_format='.0f')
    plt.title(f'Confusion Matrix: {name}')
    plt.show()
```

```python
# Apply Classifiers on Original, PCA, and LDA Data
for name, model in classifiers.items():
    print(f"\n🔹 Running {name} on Original Data")
    evaluate_model(model, X_train, X_test, y_train, y_test, name)

    print(f"\n🔹 Running {name} on PCA-Transformed Data")
    evaluate_model(model, X_pca_train, X_pca_test, y_train, y_test, f"{name} (PCA)")

    print(f"\n🔹 Running {name} on LDA-Transformed Data")
    evaluate_model(model, X_lda_train, X_lda_test, y_lda_train, y_lda_test, f"{name} (LDA)")

# Decision Boundary for KNN
def plot_decision_boundary_knn(X, y, model, title):
    h = 0.02
    cmap_light = ListedColormap(["#FFAAAA", "#AAAAFF", "#AFFFCC"])
    cmap_bold = ListedColormap(["#FF0000", "#0000FF", "#00FF00"])

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(8, 6))  # Indentation fixed
    plt.contourf(xx, yy, Z, cmap=cmap_light, alpha=0.5)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold, edgecolor="k")
    plt.xlabel("LD1")
    plt.ylabel("LD2")
    plt.title(title)
    plt.show()
```

```python
# Copy the dataset
df = anuj_df.copy()

# Target variable
target = "HOOKLOAD_AVG tonfUS"
X = df.drop(columns=[target])
y = df[target]

# Binning target variable
y_binned = pd.qcut(y, q=3, labels=["Low", "Medium", "High"])  # Create 3 bins
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y_binned)  # Convert to numerical labels

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Define classifiers
classifiers = {
    "Naive Bayes": GaussianNB(),
    "SVM": SVC(probability=True, random_state=42),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "Logistic Regression": LogisticRegression(max_iter=500, random_state=42),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42)
}

# Define the meta-classifier (for stacking)
meta_classifier = LogisticRegression()

# Create Stacking Classifier
stacking_clf = StackingClassifier(estimators=list(classifiers.items()), final_estimator=meta_classifier)

# Model Evaluation Function
def evaluate_model(model, X_train, X_test, y_train, y_test, name):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    print(f"\n🔹 Classification Report for {name}")
    print(classification_report(y_test, y_pred))

    plt.figure(figsize=(6, 4))  # Indentation fixed
    sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues',
                xticklabels=["Low", "Medium", "High"], yticklabels=["Low", "Medium", "High"])
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title(f"Confusion Matrix - {name}")
    plt.show()

# Evaluate classifiers
for name, model in classifiers.items():
    evaluate_model(model, X_train, X_test, y_train, y_test, name)

# Evaluate Stacking Classifier
evaluate_model(stacking_clf, X_train, X_test, y_train, y_test, "Stacking Classifier")
```

# SECTION 6: Perform Hyperparameter tuning using GridSearchCV

* Choose your own parameters to tune **(Don't take too many variations as time increases drastically)**
* Perform Hyperparameter tuning using GridSearchCV
* *Optional*: If you have time you can also try the Optuna library (Bayesian Optimization) for Hyperparameter tuning
* Retrain the classifiers using optimal hyperparameters
* Calculate the new metrics

```python
param_grids = {
    "Naive Bayes": {},
    
    "SVM": {
        "C": [0.1, 1, 10],
        "kernel": ["linear", "rbf"]
    },
    
    "KNN": {
        "n_neighbors": [3, 5, 7],
        "weights": ["uniform", "distance"]
    },
    
    "Logistic Regression": {
        "C": [0.1, 1, 10],
        "solver": ["lbfgs", "liblinear"]
    },
    
    "Decision Tree": {
        "max_depth": [5, 10, None],
        "min_samples_split": [2, 5, 10]
    },
    
    "Random Forest": {
        "n_estimators": [50, 100, 200],
        "max_depth": [5, 10, None]
    }
}

# Best estimators dictionary to store the tuned models
best_estimators = {}

# Perform GridSearchCV for each classifier
for name, model in classifiers.items():
    print(f"\nTuning {name} ...")
    grid = GridSearchCV(model, param_grids[name], cv=3, scoring="accuracy", n_jobs=-1)
    grid.fit(X_train, y_train)
    best_estimators[name] = grid.best_estimator_
    print(f"Best parameters for {name}: {grid.best_params_}")

# Evaluate each tuned model
for name, model in best_estimators.items():
    evaluate_model(model, X_train, X_test, y_train, y_test, f"{name} (Tuned)")

# Create Stacking Classifier with tuned estimators
meta_classifier = LogisticRegression()

stacking_clf_tuned = StackingClassifier(
    estimators=[(name, model) for name, model in best_estimators.items()],
    final_estimator=meta_classifier
)

# Evaluate the Stacking Classifier with tuned models
evaluate_model(stacking_clf_tuned, X_train, X_test, y_train, y_test, "Stacking Classifier (Tuned)")
```

### 3. **LDA-Transformed Data**

- **Performance After Tuning**:
    - Hyperparameter tuning significantly improved the performance of SVM and Random Forest.
    - Logistic Regression and Naive Bayes showed marginal improvement as they are less sensitive to hyperparameter changes.

### **Overall Observations**
- **Best Performing Algorithms**:

- **Impact of Hyperparameter Tuning**:
